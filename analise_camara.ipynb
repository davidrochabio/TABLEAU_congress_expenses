```python
import requests

import pandas as pd

import numpy as np

def expenses_cong(id_dep, years, page):

    years_lst = list(years)

    headers = {
        'accept': 'application/json',
    }

    params = (
        ('ano', years_lst),
        ('pagina', page),
        ('itens', '100'),
        ('ordem', 'ASC'),
        ('ordenarPor', 'ano'),
    )

    response = requests.get('https://dadosabertos.camara.leg.br/api/v2/deputados/' + id_dep + '/despesas',
                            headers=headers,
                            params=params)

    return response.json()


congs_df = pd.read_excel('congress_men.xlsx')

congs_ids = pd.DataFrame(congs_df['id'])

ids_lst = congs_ids['id']

congs_names = pd.DataFrame(congs_df['nome'])

names_lst = congs_names['nome']

df2_lst = []

for i in ids_lst[0:548]:
    print(i)
    df1_lst = []
    for num in range(1, 7, 1):  # iterates thru all the pages of expense results (each page has 100 results).

        expense = expenses_cong(id_dep=str(i), years=['2019', '2020'], page=str(num))  # calls the function on a congressman.

        data = expense['dados']  # key for the data of an expense.

        df = pd.DataFrame(data)  # transforms the dict into a data frame.

        df1_lst.append(df)  # appends each page df to a list of data frames

        #df2_lst.append(result_spending)

        #result_spending.to_csv('expense_' + str(i) + '.csv')
    
    result_spending = pd.concat(df1_lst, ignore_index=True)  # concatenates the data frames(pages) into 1 df.

    result_spending['id'] = i
    
    try:
        result_spending = result_spending.groupby(['id','tipoDespesa'])[['valorLiquido']].sum().reset_index()
        
        result_spending = result_spending.pivot(columns='tipoDespesa', index='id', values='valorLiquido')
    
        df2_lst.append(result_spending)

    except KeyError:
        continue
        
final = pd.concat(df2_lst)

final.to_csv('final.csv')
```
